```python
import bs4
import requests
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
```

```python
my_url = 'https://seasonskosher.com/clifton/#!/hb/c/10387-0/c/10434-10387/he/Dried-Fruit-Nuts'

# opening up connection, grabbing the page
uClient = uReq(my_url)
page_html = uClient.read()
uClient.close()
requests.get(my_url)
```

```python
# html parser
page_soup = soup(page_html,'lxml')
page_soup
```

```python
# single page crawler

my_url = 'https://www.hmart.com/groceries?p=0'
uClient = uReq(my_url)
page_html = uClient.read()
uClient.close()
page_soup = soup(page_html,'html.parser')
products = page_soup.findAll("div",{"class":"product-item-info"})

filename = "H-mart trail.csv"
f = open(filename,"w")
headers = "name, price, image, sale\n"
f.write(headers)

for product in products:
    name = product.find("a",{"class":"product-item-link"}).string.strip()
    price = product.find("span",{"class":"price"}).string
    image = product.find("span",{"class":"product-image-wrapper"}).img['src']
    sale = product.find("div",{"class":"product-price-flag"}).span
    if sale == None:
        sale = 'No Sale'
    else:
        sale = 'On Sale'

    print("name:" + name)
    print("price:" + price )
    print("image:" + image)
    print("sale:" + sale )
    
    f.write(name + "," + price+ "," + image + "," + sale + "\n")
f.close()
```

```python
# page loop

filename = "H-mart page 23.csv"
f = open(filename,"w")
headers = "name, price, image, sale\n"
f.write(headers)

for i in range(0,11):
    my_url_i = 'https://www.hmart.com/groceries/snacks?p='+str(i)
    uClient = uReq(my_url_i)
    page_html = uClient.read()
    uClient.close()
    page_soup = soup(page_html,'html.parser')
    products = page_soup.findAll("div",{"class":"product-item-info"})
    
    for product in products:
        name = product.find("a",{"class":"product-item-link"}).string.strip()
        price = product.find("span",{"class":"price"}).string
        image = product.find("span",{"class":"product-image-wrapper"}).img['src']
        sale = product.find("div",{"class":"product-price-flag"}).span
        if sale == None:
            sale = 'No Sale'
        else:
            sale = 'On Sale'

        print("name:" + name)
        print("price:" + price )
        print("image:" + image)
        print("sale:" + sale )
        
        f.write(name + "," + price+ "," + image + "," + sale + "\n")
        
f.close()
```
